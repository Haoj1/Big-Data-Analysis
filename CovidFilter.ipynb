```python
import pandas as pd
import numpy as np
```

```python
# it is a whole covid dataset by county daily
covid_df = pd.read_csv('./dataset/US_counties_COVID19_health_weather_data.csv')
```

```python
covid_df
```

```python
state_abbr = pd.read_csv('./dataset/state_abbr.csv')
state_abbr_dict = state_abbr.set_index('State').to_dict()['Abbreviation']
covid_df['state'] = covid_df['state'].map(state_abbr_dict)
```

```python
covid_df
```

The dataframe is too large, we only keep the the columns we need ['date', 'county', 'state', 'cases', 'deaths', 'stay_at_home_announced', 'stay_at_home_effective', 'total_population',
             'population_density_per_sqmi', 'date_stay_at_home_announced', 'date_stay_at_home_effective','fog','rain','snow','thunder',
            'visibility', 'max_wind_speed', 'percent_unemployed_CHR', 'percent_below_poverty', 'lat', 'lon']
The fips is to identify the county. And we have kept some weather data to exclude the possibilty that weather may affect the airlines

```python
kept_col = ['date', 'county', 'state', 'fips', 'cases', 'deaths', 'stay_at_home_announced', 'stay_at_home_effective', 'total_population',
             'population_density_per_sqmi', 'date_stay_at_home_announced', 'date_stay_at_home_effective','fog','rain','snow','thunder',
            'visibility', 'max_wind_speed', 'percent_unemployed_CHR', 'percent_below_poverty', 'lat', 'lon']
covid_clean_df = covid_df[kept_col]
covid_clean_df
```

```python
# convert the date into standard format
covid_clean_df['date'] = pd.to_datetime(covid_clean_df['date'])
covid_clean_df
```

```python
# have new columns YEAR and week
covid_clean_df['WEEK'] = covid_clean_df['date'].dt.weekofyear
covid_clean_df['YEAR'] = covid_clean_df['date'].dt.year
```

```python
covid_clean_df
```

```python
covid_clean_df.to_csv('./dataset/covid_clean_df_date', index=False)
```

We group by each county weekly using its identical fips.
For each columns after group, we use specific aggregate method 

```python
group_covid_df = covid_clean_df.groupby(['fips', 'YEAR', 'WEEK'])
```

kept_col = ['date', 'county', 'state', 'fips', 'cases', 'deaths', 'stay_at_home_announced', 'stay_at_home_effective', 'total_population',
             'population_density_per_sqmi', 'date_stay_at_home_announced', 'date_stay_at_home_effective','fog','rain','snow','thunder',
            'visibility', 'max_wind_speed', 'percent_unemployed_CHR', 'percent_below_poverty', 'lat', 'lon']

```python
group_covid_agg_df = group_covid_df.agg({
    'cases': 'last',    # we choose the case number in last day to represent the total cases a week
    'deaths': 'last',   # same as case
    'county' : 'first', 
    'state'  : 'first', 
    'stay_at_home_announced' :  lambda x: 'yes' if (x == 'yes').sum() >= 4 else 'no',
    'stay_at_home_effective' :  lambda x: 'yes' if (x == 'yes').sum() >= 4 else 'no',
    'total_population': 'first',
    'population_density_per_sqmi': 'first',
    'date_stay_at_home_announced': 'first', 
    'date_stay_at_home_effective': 'first', # the day that required to stay at home
    'percent_unemployed_CHR': 'mean',
    'percent_below_poverty': 'mean',
    'fog': 'sum',
    'rain': 'sum',
    'snow': 'sum',  
    'thunder': 'sum',               # we sum up the day with severe weather 
    'visibility': 'mean',
    'max_wind_speed': 'mean',       # we use mean value for the weather
    'lat' : 'first',                
    'lon' : 'first'
})
# group_covid_agg_df = group_covid_df.agg({'stay_at_home_effective': lambda x: 'yes' if (x == 'yes').sum() >= 4 else 'no'})
```

```python
group_covid_agg_df
```

```python
# we can find weekly increasing cases by have the diff of previous week
group_covid_agg_df['weekly_cases'] = group_covid_agg_df.groupby('fips')['cases'].diff()
group_covid_agg_df['weekly_cases'] = group_covid_agg_df['weekly_cases'].fillna(group_covid_agg_df['cases'])
```

```python
group_covid_agg_df
```

```python
# filter out invalid fips code
group_covid_agg_df.reset_index(inplace=True)
group_covid_agg_df = group_covid_agg_df[group_covid_agg_df['fips'].astype(str).str.len() == 5]
```

```python
# find the state code and county code
group_covid_agg_df['STATEFP'] = group_covid_agg_df['fips'].astype(str).str[:2]
group_covid_agg_df['COUNTYFP'] = group_covid_agg_df['fips'].astype(str).str[2:5]
```

```python
group_covid_agg_df
```

```python
group_covid_agg_df.to_csv('./dataset/covid_county_week.csv', index = False)
```
