```python
import pandas as pd
import numpy as np
```

```python
airline_df = pd.read_csv("./dataset/alljoined_airlines.csv")
```

```python
airline_df.drop(['CANCELLATION_CODE', 'DIVERTED','CARRIER_DELAY',	'WEATHER_DELAY',	'NAS_DELAY',	'SECURITY_DELAY',	'LATE_AIRCRAFT_DELAY'], axis=1, inplace=True)
```

```python
airline_df.drop(['DEP_DELAY','DEP_DELAY_NEW','DEP_DEL15','DEP_DELAY_GROUP',	'ARR_DELAY','ARR_DELAY_NEW','ARR_DEL15','ARR_DELAY_GROUP','CANCELLED'], axis=1, inplace=True )
```

```python
airline_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
airline_df.dropna(inplace=True)
airline_df
```

```python
airline_df['FL_DATE'] = pd.to_datetime(airline_df['FL_DATE'])
airline_df
```

```python
airline_df['week'] = airline_df['FL_DATE'].dt.weekofyear
```

```python
airline_df
```

```python
grouped_airline_df = airline_df.groupby(['YEAR', 'week', 'ORIGIN_AIRPORT_ID'])
grouped_airline_df = grouped_airline_df.size().reset_index(name='COUNT_FLIGHTS')
```

```python
grouped_airline_df
```

```python
main_airlines_df = grouped_airline_df[grouped_airline_df['COUNT_FLIGHTS']>=250]
```

```python
main_airlines_df
```

```python
airport_df = pd.read_csv('./dataset/airport_info.csv')
```

```python
airport_df.dropna(inplace=True)
```

```python
# airport_df['ORGIN_AIPORT_ID'] = airline_df['ORGIN_AIPORT_ID'].astype('int64')
airport_df.rename(columns={'ORGIN_AIPORT_ID': 'ORIGIN_AIRPORT_ID'}, inplace=True)
airport_df['ORIGIN_AIRPORT_ID'] = airport_df['ORIGIN_AIRPORT_ID'].astype('int64')
airport_df
```

```python
merged_airlines_df = main_airlines_df.merge(airport_df, on='ORIGIN_AIRPORT_ID', how='left')
```

```python
merged_airlines_df.dropna(inplace=True)
```

```python
merged_airlines_df.rename(columns={'Code.y':'IATA'}, inplace=True)
```

```python
merged_airlines_df
```

```python
geo_airports = pd.read_csv('./dataset/airports.csv')
geo_airports
```

```python
geo_airports_df = merged_airlines_df.merge(geo_airports, on='IATA', how='left')
```

```python
geo_airports_df.dropna(inplace=True)
```

```python
geo_airports_df['CITY'] = geo_airports_df['CITY']
geo_airports_df
```

```python
city_county = pd.read_csv('./dataset/uscities.csv')
```

```python
city_county = city_county[['city','state_id', 'county_name']]
city_county
```

```python
city_county.rename(columns={'state_id' : 'STATE', 'county_name' : 'COUNTY', 'city' : 'CITY'}, inplace=True)
```

```python
city_county['CITY'] = city_county['CITY']
city_county
```

```python
geo_airports_df = geo_airports_df.merge(city_county, on=['CITY', 'STATE'])
```

```python
geo_airports_df
```

```python
geo_airports_df.to_csv('./dataset/geo_airports_week.csv', index = False)
```
