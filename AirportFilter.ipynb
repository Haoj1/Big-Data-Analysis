```python
import pandas as pd
import numpy as np
```

```python
# read the whole csv file
airline_full_df = pd.read_csv("./dataset/alljoined_airlines.csv")
```

```python
# select the only colums we need, departure and arrival airports, flight dates, cancellation and cancellation reason
airline_df = airline_full_df[[	'YEAR',	'MONTH', 'FL_DATE',	'OP_UNIQUE_CARRIER','OP_CARRIER_FL_NUM', 'ORIGIN_AIRPORT_ID',	'DEST_AIRPORT_ID', 'CANCELLED', 'CANCELLATION_CODE']]
```

```python
#fill all NaN value with 'N'
airline_df['CANCELLATION_CODE'].fillna('N', inplace=True)
airline_df
```

The explanation for the CANCELLATION_CODE
Code,Description
"A","Carrier"
"B","Weather"
"C","National Air System"
"D","Security"

```python
# Have a new column that translate the code into explanation
codes = {"A": "Carrier", "B": "Weather", "C": "National Air System", "D": "Security", 'N' : 'No'}
airline_df['CANCELLATION_DESCRIPTION'] = airline_df['CANCELLATION_CODE'].map(codes)
```

```python
airline_df
```

```python
# drop all invalid data
airline_df.dropna(inplace=True) 
```

```python
# we transform the the flight date to standard expression so that we could extract week from the date
airline_df['FL_DATE'] = pd.to_datetime(airline_df['FL_DATE'])
airline_df
```

```python
# have a new columns week for each flight, we will use Year and week to categorize the flights number
airline_df['WEEK'] = airline_df['FL_DATE'].dt.weekofyear
```

```python
airline_df
```

```python
# We group by YEAR', 'WEEK', 'ORIGIN_AIRPORT_ID', which means that we focus on the flights number of a airport weekly instead of a single day. Since airlines differ
# everyday, the daily airlines number would change significantly. 
# And we only count the depature flights, according to an assumption that a flight always travels between two airports. There must be an arrival after a departure.
grouped = airline_df.groupby(['ORIGIN_AIRPORT_ID', 'YEAR', 'WEEK'])
# group the weekly cancellation using sum
grouped_airline_df = grouped.agg({'CANCELLED': 'sum'})
# count the number of flight in a specific airport in a specific year and week
grouped_airline_df['COUNT_FLIGHTS'] = grouped.size()
# count the numbers of reason weekly, it could be weather, carrier itself, National Air System, Security
grouped_airline_df['WEATHER_CANCELLED'] = grouped['CANCELLATION_DESCRIPTION'].apply(lambda x : (x == 'Weather').sum())
grouped_airline_df['NAS_CANCELLED'] = grouped['CANCELLATION_DESCRIPTION'].apply(lambda x : (x == 'National Air System').sum())
grouped_airline_df['SEC_CANCELLED'] = grouped['CANCELLATION_DESCRIPTION'].apply(lambda x : (x == 'Security').sum())
grouped_airline_df['CARRIER_CANCELLED'] = grouped['CANCELLATION_DESCRIPTION'].apply(lambda x : (x == 'Carrier').sum())
```

```python
grouped_airline_df
```

```python
# we ignore some small airports that avg flight number < 100
main_airlines_df = grouped_airline_df[grouped_airline_df.groupby('ORIGIN_AIRPORT_ID')['COUNT_FLIGHTS'].transform('mean') >= 100]
# main_airlines_df = grouped_airline_df[grouped_airline_df['COUNT_FLIGHTS'] >= 100]
```

```python
main_airlines_df
```

```python
main_airlines_df = main_airlines_df.reset_index()
```

We want to know where the airport is, so we have another dataset that contain the geo locations

```python
# we have an identified code for each airport
airport_df = pd.read_csv('./dataset/airport_info.csv')
airport_df
```

```python
# airport_df['ORGIN_AIPORT_ID'] = airline_df['ORGIN_AIPORT_ID'].astype('int64')
airport_df.dropna(inplace=True)
airport_df.rename(columns={'ORGIN_AIPORT_ID': 'ORIGIN_AIRPORT_ID'}, inplace=True)
airport_df['ORIGIN_AIRPORT_ID'] = airport_df['ORIGIN_AIRPORT_ID'].astype('int64')
airport_df
```

```python
# merge the flights number dataset with the identified code data
merged_airlines_df = main_airlines_df.merge(airport_df, on='ORIGIN_AIRPORT_ID')
```

```python
merged_airlines_df.dropna(inplace=True)
```

```python
merged_airlines_df.rename(columns={'Code.y':'IATA'}, inplace=True)
```

```python
merged_airlines_df
```

```python
# Now we have the IATA code, but it is not enough, we have to find the location that is with identical IATA code
geo_airports = pd.read_csv('./dataset/airports.csv')
geo_airports
```

The dataset contains the city and state, but without the county which is the exact we want, so we keep the LATITUDE	LONGITUDE to find which county it is exactly located

```python
geo_airports_df = merged_airlines_df.merge(geo_airports, on='IATA', how='left')
```

```python
geo_airports_df.dropna(inplace=True)
```

```python
geo_airports_df['CITY'] = geo_airports_df['CITY']
geo_airports_df
```

We have a completed dataframe that groupby airport weekly and contains the geo location

```python
#export to csv
geo_airports_df.to_csv('./dataset/geo_airports_week.csv', index = False)
```
