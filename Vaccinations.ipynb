```python
import pandas as pd
import numpy as np
```

```python
covid_df = pd.read_csv('./dataset/us-counties.csv')
```

```python
covid_df
```

```python
covid_df.dropna(inplace=True)
covid_df['fips'] = covid_df['fips'].astype(int)
covid_df
```

```python
covid_df['STATEFP'] = covid_df['fips'].astype(str).str[:2]
covid_df['COUNTYFP'] = covid_df['fips'].astype(str).str[2:5]
covid_df
```

```python
covid_df['date'] = pd.to_datetime(covid_df['date'])
covid_df['WEEK'] = covid_df['date'].dt.weekofyear
covid_df['YEAR'] = covid_df['date'].dt.year
```
