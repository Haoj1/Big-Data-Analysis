```python
import pandas as pd
import geopandas as gpd
```

```python
county_geo_gdf = gpd.read_file('./dataset/cb_2017_us_county_5m/cb_2017_us_county_5m.shp')
```

```python
county_geo_gdf.crs = {'init': 'epsg:3857'}
```

```python
from shapely.geometry import Point
```

```python
airline_df = pd.read_csv('./dataset/geo_airports_week.csv')
airline_df.head()
```

```python
points  = [Point(x, y) for x, y in zip(airline_df['LONGITUDE'], airline_df['LATITUDE'])]
# Convert the radius from miles to degrees
# 1 degree = 69.172 miles
radius = 20 / 69.172 
circles = [Point(p.x, p.y).buffer(radius) for p in points]
airline_gdf = gpd.GeoDataFrame(airline_df, geometry=circles)
airline_gdf.crs = {'init': 'epsg:4326'}
airline_gdf
```

```python
matching_counties = []
airline_county_gdf = gpd.sjoin(airline_gdf, county_geo_gdf, how='left', op='intersects')
```

```python
airline_county_gdf
```

```python
airline_county_gdf.to_csv('./dataset/airline_county_gdf.csv', index = False)
```

```python
airline_county_gdf['Matching_County'] = airline_county_gdf['NAME']
airline_county_gdf
```

```python
covid_df = pd.read_csv('./dataset/covid_county_week.csv')
covid_df
```

```python
geometry = [Point(x, y) for x, y in zip(covid_df['lon'], covid_df['lat'])]
covid_gdf = gpd.GeoDataFrame(covid_df, geometry=geometry)
covid_gdf
```

```python
# Create a list to store the matching counties
matching_counties = []

# Use spatial join to join the two GeoDataFrames and find the matching counties
result = gpd.sjoin(covid_gdf, county_geo_gdf, how='left', op='within')
matching_counties = result['NAME'].tolist()
covid_gdf['Matching_County'] = matching_counties
```

```python
covid_gdf.rename(columns={'COUNTY' :'county', 'STATE' : 'state'}, inplace=True)
airline_county_gdf.rename(columns={'COUNTY' :'county', 'STATE' : 'state'}, inplace=True)
covid_gdf.drop(['county'], axis=1, inplace=True)
```

```python
covid_gdf.to_csv('./dataset/covid_gdf.csv', index = False)
```

```python
covid_airline_gdf = airline_county_gdf.merge(covid_gdf, on = ['Matching_County','state', 'YEAR', 'week'])
```

```python
covid_airline_gdf.to_csv('./dataset/covid_airline.csv', index = False)
```
